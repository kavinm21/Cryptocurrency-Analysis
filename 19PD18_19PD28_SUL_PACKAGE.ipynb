```python
# Removing unwanted Columns
df_bit=df_bit.drop(['Name','Symbol','Date'],axis=1)
```

```python
# Ordering the columns manually
columns_titles = ["High","SNo","Low","Open","Close","Volume","Marketcap"]
df_bit=df_bit.reindex(columns=columns_titles)
```

```python
df_bit.dtypes
```

```python
df_bit
```

```python
# Splitting data into input and output
y_bit=df_bit.iloc[:,0].values
X_bit=df_bit.iloc[:,1:7].values
```

```python
np.mean(y_bit)
```

### Linear Regression Model 

```python
scaler_bit = StandardScaler()
X_bit = scaler_bit.fit_transform(X_bit)
y_bit = (y_bit-y_bit.mean())/y_bit.std()
```

```python
X_bit_train,X_bit_test,y_bit_train,y_bit_test = train_test_split(X_bit,y_bit,test_size=0.33,random_state=0)
```

```python
linearRegr_bit = LinearRegression()
```

```python
linearRegr_bit.fit(X_bit_train, y_bit_train)
```

```python
y_bit_pred_linear=linearRegr_bit.predict(X_bit_test)
```

```python
mean_squared_error(y_bit_test,y_bit_pred_linear)
```

```python
mean_absolute_error(y_bit_test,y_bit_pred_linear)
```

It acts as the best model for the bitcoin data.

 Graph between the predicted high and actual high value of given days.\
 Red represents the given data.\
 Blue represents the predicted data.

```python

fig1_bit_linear = px.scatter(df_bit, y=y_bit_test, color_discrete_sequence=['red'])
fig2_bit_linear = px.scatter(df_bit, y=y_bit_pred_linear, color_discrete_sequence=['blue'])
fig3_bit_linear = go.Figure(data=fig1_bit_linear.data + fig2_bit_linear.data)
fig3_bit_linear.update_layout(title='Linear Regression Model',xaxis_title="Order/SNo", yaxis_title="High Value of Day")
fig3_bit_linear.show()
```
